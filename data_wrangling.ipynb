```python
import pandas as pd

# create DataFrame from RewardsData CSV
rewards_df = pd.read_csv('RewardsData.csv')

# view first 20 records
rewards_df.head(20)
```

```python
# return boolean values (T or F)indicate whether the element is missing aka NaN/None/NaT (T) or has a value (F)
rewards_df.isna()
```

```python
# count number of missing values for each column
missing_value_counts = rewards_df.isna().sum()
missing_value_counts
```

```python
missing_value_perentages = missing_value_counts/len(rewards_df) * 100
missing_value_perentages
```

```python
# last seen has low percentage of missing values, potential candidate to fill in missing values with a default value
rewards_df['Last Seen'] # see dtype
```

```python
rewards_df['Last Seen'] = pd.to_datetime(rewards_df['Last Seen'])
```

```python
rewards_df['Last Seen']
```

```python
rewards_df['Last Seen'].max()
```

```python
rewards_df['Last Seen'].min()
```

```python
# replace any missing value with specified value
rewards_df['Last Seen'].fillna(value=rewards_df['Last Seen'].min(), inplace=True)
```

```python
missing_value_counts = rewards_df.isna().sum()
missing_value_counts
```

```python
# drop records with missing values
rewards_df_sample = rewards_df.dropna()
rewards_df_sample
```

```python
# confirm no missing values
missing_value_counts_sample = rewards_df_sample.isna().sum()
missing_value_counts_sample
```

```python
# drop the Tags column from rewards_df_sample
rewards_df_sample = rewards_df_sample.drop(columns='Tags')
rewards_df_sample
```

```python
# get list of unique Cities
city_unique_values = rewards_df_sample['City'].unique()

#sort values
city_unique_values.sort()
city_unique_values # we see alot of winston salem in different formats > problem
```

```python
# replae variations of Winston-Salem with the correct format
rewards_df_sample['City'].replace(to_replace=['Winston Salem', 'Winston Salem ',
       'Winston salem', 'Winston salem ', 'Winston-Salem',
       'Winston-Salem ', 'Winston-Salem, NC', 'Winston-salem',], value="Winston-Salem", inplace=True)
```

```python
# remove all leading and trailing spaces from City column
rewards_df_sample['City'] = rewards_df_sample['City'].str.strip()
```

```python
# fix any upper-lower casing issues present in City list
rewards_df_sample['City'] = rewards_df_sample['City'].str.title()
```

```python
# get list of unique Cities
city_unique_values = rewards_df_sample['City'].unique()

#sort values
city_unique_values.sort()
city_unique_values 
```

```python
# repeat for State and Zip
# get list of unique States
state_unique_values = rewards_df_sample['State'].unique()

#sort values
state_unique_values.sort()
state_unique_values 
```

```python
# get list of unique States
state_unique_values = rewards_df_sample['State'].unique()

#sort values
state_unique_values.sort()
state_unique_values 
```

```python
# include full State name along with abbrev in dataset > use merging
# convert us_state_to_abbrev dict to DataFrame for merging
states_df = pd.DataFrame(list(us_state_to_abbrev.items()), columns=['State Name', 'State'])
states_df.head(3)
```

```python
# merge original DataFrame with States DataFrame > inner/left/right/outer
rewards_df_sample = pd.merge(rewards_df_sample, states_df, on='State', how='left')
rewards_df_sample
```

```python
# rename 'State' column to 'State Abbreviation'
rewards_df_sample.rename(columns={'State': 'State Abbreviation'}, inplace=True)
rewards_df_sample.head(3)
```

```python
# move State Name column to after State Abbrev column
# step 1: store column data and remove it from original position using pop() method
state_name_col = rewards_df_sample.pop('State Name') #remove a column (or row, in Series) from a DataFrame and return it as a separate object.
# step 2: insert the column back into the DF at the new position -> 1st arg is the new position, 2nd arg is the new column name, 3rd arg is the data of the column
rewards_df_sample.insert(3, 'State Name', state_name_col)
rewards_df_sample.head(3)
```

```python
# generate a list of unique values for Zip column from state column
state_unique_values = rewards_df_sample['Zip'].unique()
state_unique_values
```

```python
len(state_unique_values) # no. of unique values in array
```

```python
# truncate all zip codes with more than 5 digits (e.g. 27707-1634) down to five
rewards_df_sample['Zip'] = rewards_df_sample['Zip'].str[:5]
```

```python
# create a booleans index to check for zip codes with length of five digits
zip_code_boolean_index = rewards_df_sample['Zip'].str.len() == 5 #In Pandas, Boolean indexing is used to filter rows to give T/F based on criteria
zip_code_boolean_index
```

```python
rewards_df_sample = rewards_df_sample[zip_code_boolean_index] # drop records that have zip code length of more than 5
zip_unique_values = rewards_df_sample['Zip'].unique() # get list of unique values for Zip column
zip_unique_values.sort()
zip_unique_values
```
