```python
import pandas as pd

#create DataFrame from CleanRewardsData CSV
rewards_df_clean = pd.read_csv('CleanRewardsData.csv')
rewards_df_clean
```

```python
# group records by state name as it has many repeats
state_groups = rewards_df_clean.groupby('State Name')
state_groups
```

```python
# output the sum aggregation values for numeric columns
state_groups.sum(numeric_only=True)
```

```python
#output multiple aggregation values for 'Points Spent' column
state_groups['Points Spent'].agg(['count', 'sum', 'mean', 'max', 'min'])
```

```python
#output multiple aggregation values for multiple columns
state_groups.agg({'Total Points Earned': ['sum','max','min'], 'Points Spent':['sum','max','min']})
```

```python
# group records by city and state name
city_state_groups = rewards_df_clean.groupby(['City','State Name'])
city_state_groups.sum(numeric_only=True)
```
