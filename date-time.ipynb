```python
# initialize pandas timestamp variables for birthdate and today EXAMPLE
birthdate = pd.to_datetime('1992-01-31')
today = pd.to_datetime('2024-01-15')

# extract year values from dates using Python's datetime instance attributes
birthdate_year = birthdate.year
today_year = today.year

#calculate the year difference
year_diff = today_year - birthdate_year

age = year_diff - ((today.month, today.day) < (birthdate.month, birthdate.day))
print(age)
```

```python
# define a function using input
def calculateAge(birthdate):
    today = pd.to_datetime('today')

    # extract year values from dates using Python's datetime instance attributes
    birthdate_year = birthdate.year
    today_year = today.year

    #calculate the year difference
    year_diff = today_year - birthdate_year

    age = year_diff - ((today.month, today.day) < (birthdate.month, birthdate.day))

    return age
```

```python
# apply calculateAge() function to update 'User Age' column values
rewards_df_sample['User Age'] = rewards_df_sample['Birthdate'].apply(calculateAge)
rewards_df_sample.head(3)
```
