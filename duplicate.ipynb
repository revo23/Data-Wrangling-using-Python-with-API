```python
import pandas as pd

#create DataFrame from RewardsData CSV
rewards_df_sample = pd.read_csv('WrangledRewardsData.csv')
rewards_df_sample
```

```python
# check for duplicates
duplicates = rewards_df_sample.duplicated()
duplicates
```

```python
# get the count of duplicate records
duplicates.sum()
```

```python
# remove duplicates
rewards_df_clean = rewards_df_sample.drop_duplicates()

# confirm duplicates haev been removed
duplicates = rewards_df_clean.duplicated()
duplicates.sum()
```

```python
# export dataframe to CSV file
rewards_df_clean.to_csv("CleanRewardsData.csv")
```

```python
# remove duplicates based on specific columns
# create sample dataframe with duplicate values within columns
data = {'A':[1,2,2,3,3,3], 'B':[1,1,2,2,3,3]}
df = pd.DataFrame(data)
df
```

```python
# remove duplicates based on subset of columns
df_unique_subset = df.drop_duplicates(subset=['A'], keep='first') # keep first occurence
df_unique_subset
```
